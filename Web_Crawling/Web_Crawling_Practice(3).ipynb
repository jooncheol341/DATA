## HTML
- Hypertext Markup Language
- 웹사이트에서 화면에 표시되는 정보를 약속한 것
- markup : 꺽쇠( < , >)로 이뤄진 태그를 사용하는 규격

## XML
- extensible 
- HTML의 한계를 극복할 목적으로 만들어진 다목적 마크업 언어
- XML는 HTML처럼 데이터를 보여줄 목적이 아닌, 데이터를 저장하고 전달할 목적으로만 만들어짐
- XML : 태그는 HTML 태그처럼 미리 정의되어 있지 않고, 사용자가 직접 정의 할 수 있음.

```python
import requests
url = "http://www.kma.go.kr/wid/queryDFSRSS.jsp?zone=1111061500"
data = requests.get(url).text
data
```

```python
#xml - > dic
import xmltodict as xtd

# parse -> 해석, 해독, 변환의 의미로 들어있는 정보는 달라지지 않고, 형식만 바꾸는 것
change = xtd.parse(data)
change
```

```python
type(change)
```

```python
change['rss']['channel']['pubDate']
```

```python
data1 = change['rss']['channel']['item']['description']['body']['data']
data1[0]['wfKor']
```

```python
data1
```

```python
import pandas as pd

pd.DataFrame(data1).to_csv('weather.csv', index= False)

#한글이 꺠져서 저장됨
#엑셀로 열었을 떄, 메모장으로 열었을 떄 안 꺠짐 -> 데이터가 손상된 것은 아님
#저장하는 방식과 읽어들이는 방식이 다를 때 발생
#해결 : 저장하는 방식과 읽어들이는 방식을 맞춰주면 됨
```

```python
pd.DataFrame(data1).to_csv('weather1.csv', index= False, encoding = 'utf_8_sig')
```

```python
import requests
url1 = "https://news.google.com/rss/search?q=반포동&hl=ko&gl=KR&ceid=KR%3Ako"
data = requests.get(url1).text
data
```

```python
import xmltodict as xtd
data1 = xtd.parse(data)
news_list = data1['rss']['channel']['item']
news_list
```

```python
import pandas as pd
df = pd.DataFrame(news_list)
df
```

```python
df.to_csv('newsample.csv', index = False, encoding = 'UTF-8')
```

```python
import xmltodict
import pandas as pd
def news_get(keyword):
    url = f"https://news.google.com/rss/search?q={keyword}&hl=ko&gl=KR&ceid=KR%3Ako"
    data = requests.get(url).text
    news_data = xmltodict.parse(data)
    news_list = news_data['rss']['channel']['item']
    df = pd.DataFrame(news_list)
    return df.to_csv('news_sample.csv', index= False, encoding='UTF-8')
```
