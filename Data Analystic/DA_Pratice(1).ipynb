```python
all_list = [] #   

all_list.append({'Date' : '2023-01-01', 'DayOfWeek' : 'Sun'}) # row    dictionary   append  . 
all_list.append({'Date' : '2023-01-02', 'DayOfWeek' : 'Mon'})
all_list.append({'Date' : '2023-01-03', 'DayOfWeek' : 'Tue'})

all_list
```

```python
name = ['jo', 'kim', 'lee', 'park', 'choi']
age = [20, 30, 40, 50, 60]

DataSet = list(zip(name, age)) #     .
```

```python
pd.DataFrame(DataSet, columns = ['Name', 'Age']) # DataFrame(, columns( column    .))
```
