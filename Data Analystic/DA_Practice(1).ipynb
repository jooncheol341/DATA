## DataFrame ( 표의 형식 )
- 데이터가 틀에 맞춰서 저장된 형태 
- 테이블, 스프레드시트, 엑셀, 등 분야마다 다양하게 불린다. 
- 가로는 행( row ), 세로는 열( column )이라고 불린다. column은 보통 해당 프레임의 변수들의 이름의 모음
- 한 row 안의 데이터는 **Dictionary**처럼 저장 되어있다. 이것을 이해해야 분석이 가능하다. row에서 해당 컬럼이 key, 데이터가 value라고 볼 수 있다.
- 한 column 안의 데이터는 **list**처럼 저장되어있다. column안의 데이터들이 리스트처럼 나열되어있다. 
- 한 DataFrame은 row들이 **list**처럼 저장 되어있다. 


```python
import pandas as pd # 모듈 불러오기
```

```python
all_list = [] # 빈 리스트를 생성

all_list.append({'Date' : '2023-01-01', 'DayOfWeek' : 'Sun'}) # row에 들어갈 데이터 정보를 dictionary로 만들어주고 리스트에 append를 통해 넣어준다. 
all_list.append({'Date' : '2023-01-02', 'DayOfWeek' : 'Mon'})
all_list.append({'Date' : '2023-01-03', 'DayOfWeek' : 'Tue'})

all_list
```

```python
pd.DataFrame(all_list) # 만든 리스트를 데이터프레임화 시킬수 있다. 
# 딕셔너리의 key들이 column이 되었고 value들이 row가 되었다. 
```

```python
name = ['jo', 'kim', 'lee', 'park', 'choi']
age = [20, 30, 40, 50, 60]

DataSet = list(zip(name, age)) # 각 요소에 맞게 압축한 것이다.
```

```python
pd.DataFrame(DataSet, columns = ['Name', 'Age']) # DataFrame(리스트, columns(사용하면 column의 이름을 넣어줄 수 있다.))
```
