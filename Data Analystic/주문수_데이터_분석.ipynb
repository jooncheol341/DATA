```python
import pandas as pd # 라이브러리 불러오기
```

```python
# 고객 데이터, 이름, 성별 불러오기

customer_master = pd.read_csv('/content/customer_master.csv')
customer_master.head()
```

```python
item_master = pd.read_csv('/content/item_master.csv')
item_master
```

```python
transaction_1 = pd.read_csv('/content/transaction_1.csv')
transaction_1.head()
```

```python
transaction_detail_1 = pd.read_csv('/content/transaction_detail_1.csv')
transaction_detail_1.head()
```

```python
# transaction_1과 transaction_2는 동일한 데이터이기 때문에 합쳐준다.
transaction_2 = pd.read_csv('/content/transaction_2.csv')
transaction = pd.concat([transaction_1, transaction_2], ignore_index = True)
transaction.head()
```

```python
# 올바르게 세로로 유니온이 됬는지 확인하기 위해 len 함수를 사용하여 개수를 파악

print(len(transaction_1))
print(len(transaction_2))
print(len(transaction))
```

```python
# transaction_detail도 마찬가지이므로 세로로 유니온을 해준다.

transaction_detail_2 = pd.read_csv('/content/transaction_detail_2.csv')
transaction_detail = pd.concat([transaction_detail_1, transaction_detail_2], ignore_index = True)
transaction_detail.head()
```

```python
# transaction과 transaction_detail은 transaction_id가 공통으로 들어가 있으며 각 transaction_id에 따른 구체적인 구매 사항이 적혀있으므로
# transaction_id를 기준으로 두 데이터 프레임을 결합하면 될 것 같다.
```

```python
# transaction 데이터 프레임에서 필요한 컬럼만 추출하여 merge를 시켜 불필요한 중복을 없애준다!

join_data = pd.merge(transaction_detail, transaction[["transaction_id", "payment_date", "customer_id"]],
                     on = "transaction_id", how = "left")
join_data.head()
```

```python
print(len(transaction_detail))
print(len(transaction))
print(len(join_data))
```

```python
# join_data에 master 컬럼들을 전부 붙이기

join_data = pd.merge(join_data, customer_master, on = 'customer_id', how = 'left')
join_data = pd.merge(join_data, item_master, on = "item_id", how = "left")
join_data.head()
```

```python
# 데이터 프레임들을 합치면서 총 price 컬럼이 사라졌으므로 계산을 통해 새로운 컬럼을 생성해주어야한다.

join_data['price'] = join_data["quantity"] * join_data["item_price"]
join_data[["quantity", "item_price", "price"]]
```

```python
# 데이터를 가공했기에 총 매출이 맞는지 확인하는 검산 작업이 꼭 필요하다.

join_data["price"].sum() == transaction["price"].sum()
```

```python
# 결측치 파악 및 총 데이터 개수 파악해보기

join_data.isnull().sum()
```

```python
# 시계열 분석을 위해 payment_time이 현재 object 형식이므로 datetime으로 바꿔줄 예정
```

```python
join_data["payment_date"] = pd.to_datetime(join_data["payment_date"])
join_data["payment_month"] = join_data["payment_date"].dt.strftime("%Y%m")
join_data[["payment_date", "payment_month"]].head()
```

```python
# groupby를 활용한 월별 매출

join_data.groupby("payment_month").sum()["price"]
```

```python
# groupby를 활용하여 월 별 각 아이템의 매출을 살펴보자

join_data.groupby(["payment_month", "item_name"]).sum()[['price', 'quantity']]
```

```python
# groupby 함수를 사용하여 확인하기 보다는 pivot_table을 활용하여 조금 더 편하게 보는 것이 좋아보인다.

# E의 매출이 가장 높게 나오지만 구매하는 개수는 A 제품이 더 많은 것을 확인할 수 있었다.

# 비싸고 좋다고 많은 사람들이 선택하는 것은 아닌 것 같다는 것 또한 확인 할 수 있었다.

pd.pivot_table(join_data, index = 'item_name', columns = 'payment_month', values = ['price', 'quantity'], aggfunc = 'sum')
```

```python
# 시각화를 위해 상품별 매출 추이를 알아보자~

graph_data = pd.pivot_table(join_data, index = 'payment_month', columns = 'item_name', values = 'price', aggfunc = 'sum')
graph_data
```

```python
# matplotlib 라이브러를 활용하여 시각화를 해보자

import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(list(graph_data.index), graph_data["PC-A"], label = 'PC-A')
plt.plot(list(graph_data.index), graph_data["PC-B"], label = 'PC-B')
plt.plot(list(graph_data.index), graph_data["PC-C"], label = 'PC-C')
plt.plot(list(graph_data.index), graph_data["PC-D"], label = 'PC-D')
plt.plot(list(graph_data.index), graph_data["PC-E"], label = 'PC-E')
plt.legend()
```

시각화를 통해 제품 E의 매출이 다른 제품들에 비해 많은 매출을 가져다 주는 것을 확인할 수 있었다.
