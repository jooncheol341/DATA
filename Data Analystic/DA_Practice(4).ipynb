# DataFrame 합치기

# concat ( 데이터를 연결해서 붙인다.)

```python
import pandas as pd
```

```python
dic = { '가' : ['가1', '가2', '가3'],
        '나' : ['나1', '나2', '나3'],
        '다' : ['다1', '다2', '다3'] }
ex1 = pd.DataFrame(dic)
ex1
```

```python
dic1 = { '나' : ['나1', '나2'],
         '다' : ['다1', '다2'],
         '라' : ['라1', '라2'],
         '마' : ['마1', '마2'] }
ex2 = pd.DataFrame(dic1)
ex2
```

```python
# 구조가 맞지 않아도 concat이 가능하다.


pd.concat([ex1, ex2])
```

```python
# axis = 0 위 아래로 붙여진다. axis = 1 옆으로 붙는다. 

pd.concat([ex1, ex2], axis = 1)
```

```python
# 중복된 column만 붙이고 싶을 때는 join
# inner 겹치는 value만 보여준다. 

pd.concat([ex1, ex2], join = 'inner')
```

# Merge ( 병합 )

```python
# inner 교집합인 내용만 추출해서 수행

pd.merge(ex1, ex2)
```

```python
# outer은 합집합

pd.merge(ex1, ex2, how = 'outer')
```

# Join

```python
# join 사용 시 데이터프레임을 지정해주어야한다. 
# column이 중복되면 안된다. 

# ex1.join(ex2)
```

```python
ex1 = ex1.set_index('나')
ex1
```

```python
ex2 = ex2.set_index('다')
ex2
```

```python
ex1.join(ex2)
```

# 데이터 분석 방법론

- 전통적인 데이터 분석 방법론 : KDD, CRISP-DM


#### CRISP-DM
- Cross-Industry Standard Process - Data Mining

#### 1. Business Understading 
- 다양한 분야의 목적과 요구 사항을 이해하는 단계
- 업무 목적, 상황 파악, 목표 설정, 프로젝트 계획 수립
- 해당 분야의 종사자들이 원하는 방향, 제조업, 숙박업??? 이 사람들이 원하는 것이 무엇일까, 
- 분석의 방향성 및 목표 설정을 정해야한다.


#### 2. Data Understanding 
- 데이터를 수집하고, 데이터의 속성을 이해하는 단계
- 데이터 기술적인 분석, EDA, 데이터 품질 확인

#### EDA( Exploratory Data Analysis )
- 시각화를 통해서 데이터에서 유효한 패턴(insight)을 찾는 과정
- 그래프, 통계적인 방법 -> 자료를 직관적으로 바로보는 과정 
- Business Understanding 단계에서 설정한 방향성이 맞는지 여부 확인

1. 데이터 출처와 주제 대한 이해
2. 데이터 크기 확인
3. 데이터 구성 요소 확인
4. 탐색한 데이터의 시각화

#### 데이터 품질
- 결측치
- 일관성
- 형태
- 정보

### Data Preparation

- 다양한 분석 기법에 알맞은 데이터를 준비하는 단계
- 모델링을 대비하는 단계 
- 알고리즘에 넣을 데이터를 '적절하게' 준비 
- train / test split

#### Machine Learning
- 학습 방식에 따라 지도 학습, 비지도 학습으로 나뉨
- 데이터셋 분리(split): 학습용(train) / 시험용(test)

### Modeling

- 다양한 모델링 기법과 알고리즘 선택하는 단계
- 모델의 성능 평가, overfitting 되었는지 확인

#### Overfitting(과적합, 너무 과하게 적합되어있다.), Underfitting(과소적합)

- 훈련 데이터에 너무 최적화되어있어서 시험 데이터를 알맞지 못하게 평가하게된다.

#### Model Planning

- 속도, 예측력, 해석력을 고려해서 알고리즘 후보군 선정
- 내부 계산이 복잡할수록 속도가 느려지고, 예측력은 좋아지고, 해석력은 떨어지게된다. 
- 기술통계 분야 알고리즘은 속도와 해석력이 우수함

#### Model Building

- 채점할 때 관점에 따라 평가지표가 달라짐
- ex) 정답: 1, 예측: 2 --> 오차: 1
- ex) 정답: 1001, 예측: 1002 --> 오차: 1

### Evaluation

- 결과 토의, 비즈니스 이해에서 설정한 목표나 방향성에 부합하게 된건지 

### Deployment

- 완성된 모델을 실제 업무에 적용하기 위한 단계
